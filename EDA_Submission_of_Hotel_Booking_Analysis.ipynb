# **Project Summary -**

The aim of this project of Exploratory Data Analysis on Hotel Booking investigates cancellations to find its patterns and to suggest measures that can implemented to reduce the cancellations and secure more revenue. This given dataset covers booking informations for the city hotel and resort hotel including informations such as when the booking was made, length of stay, the number of adults and children.

The project is done to find some suggestion measures that can be implemented to reduce cancellations and secure revenue.

For example: Hotels can offer discounts for the customers to stay longer with them. Hotels can also offer extra perks like free breakfast/parking for customers those book hotels directly without using third party applications.

This project involves most of the steps in exploration starting from data collection,data cleaning (removing duplicates/null values), then identifying process related variables to analyse followed by Visualization to come up with results.

This EDA is done through three forms of analysis such as ;

Univariate Analysis : Analysing data with respect to one variable.

Bivariate Analysis : Analysing and comparing is done between two variables to study relationships between them.

Multivariate Analysis : Similiar to bivariate but here more than two variables are involved.

This project concludes that by analysing hotel booking data and understanding cancellation patterns and to reduce the cancellations and increase the revenue.
