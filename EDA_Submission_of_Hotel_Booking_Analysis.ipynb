### Variables Description

1.**hotel** : Hotel type (City hotel or Resort hotel)

2.**is_canceled** : It indicates if the booking was cancelled(1) or not(0)

3.**lead_time** : Number of days that elapsed between the entering date of the booking into PMS and the arrival date

4.**arrival_date_year** : Year of customer arrival date

5.**arrival_date_month** : Month of customer arrival date

6.**arrival_date_week_number** : Week number of customer arrival date

7.**arrival_date_day_of_month** : Day of customer arrival date

8.**stays_in_weekend_nights** : Number of weekend nights  the guest booked or stayed at the hotel

9.**stays_in_week_nights** :  Number of week nights  the guest booked or stayed at the hotel

10.**adults** : Number of adults

11.**children** : Number of children

12.**babies** : Number of babies

13.**meal** : Kind of meal customers opted for

14.**country**  : Country code (from where the customers are)

15.**market_segment**  : which market segment the custimers belong to

16.**distribution_channel**  : Booking channel (how customers accessed the stay).Here 'Direct' means direct booking, 'TA' means Travel Agents and 'TO' means Tour Operators.

17.**is_repeated_guest** : Guest coming for the hotel is first time (0) or not (1)

18.**previous_cancellations**  : was there any cancellations done before

19.**previous_bookings_not_canceled**  : Number of the previous bookings not cancelled by the customers prior to current booking

20.**reserved_room_type**  : Type of rooms reserved by the customer which is represented via codes

21.**assigned_room_type** : Type of rooms allocated to the customer which is also represented via codes

22.**booking_changes**  : Number of  changes made to the bookings from the moments booking started to till finishing

23.**deposit_type** : Indication of customer those who paid any deposit or not

24.**agent** : ID of the Agent who made the bookings

25.**company** : ID of the company which made the booking or responsible for the payment of booking

26.**days_in_waiting_list** : Number of days that the bookings was in waiting before gets confirmed

27.**customer_type** : Type of customer which is classified into some types

28.**adr** : Average Daily Rate as defined by dividing all the sum of lodging transactions by the total number of staying nights

29.**required_car_parking_spaces** : If the customer need car parking or not ?

30.**total_of_special_requests** : Number of additional special request made by the customer like extra bed, good view from the room etc,..

31.**reservation_status** : Current status of reservations based on some categories like Cancelled, Check-out, No Show

32.**reservation_status_date** : Date of current updated status of reservations

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Creating a duplicate of the original dataset to perform analysis
hba_df1 = hba_df.copy()
````

````python
# replacing the null values with 0 in the children column for those who doesn't with children
# likewise replacing the null values with 0 in the company and agents column cause some customers booked on their own

hba_df1['children'].fillna(0,inplace=True)
hba_df1['company'].fillna(0,inplace=True)
hba_df1['agent'].fillna(0,inplace=True)

# replacing the null values with 'others' in the country column for those who doesn't mentioned country

hba_df1['country'].fillna('others',inplace=True)
````

````python
hba_df1.isnull().sum()
````

````python
#decided to drop a "company" column because it has more null values

hba_df1.drop(['company'], axis=1, inplace=True)
````

````python
hba_df1.shape
````

````python
#dropping rows which doesn't have booking on that day on the basis of no adult, children and babies in that row

no_booking_day = hba_df1[hba_df1['adults']+hba_df1['children']+hba_df1['babies']==0]
hba_df1.drop(no_booking_day.index,inplace=True)
````

````python
hba_df1.shape
````

````python
#adding some combined columns to make analysis even more clear

#creating 'total_people' column with addition of all adults, children and babies
#creating 'total_stay' column by adding all kind of stays

hba_df1['total_people']=hba_df1['adults']+hba_df1['children']+hba_df1['babies']

hba_df1['total_stay']=hba_df1['stays_in_week_nights']+hba_df1['stays_in_weekend_nights']

````
